## Contar valores

dataframe.serie.value_count

Parametros ultiles:
* value_counts(ascending=True)
* value_counts(dropna=True)
  
### Ejemplos:

Cuantos vendedores unicos tenemos?

````python
tabla.Vendedor.value_counts()
````

Como podemos ver en el ejemplo nos devuelve una lista de vendedores, con la cantidad de veces que se encuentran repetidos en el dataframe. Se muestran de forma descendente porque asi esta definido por defecto. En siguiente ejemplo veremos como lista las regiones y cuantas veces se repiten las zonas.

````python
tabla.Región.value_counts(ascending=True, dropna=False)
````

Vamos a ver los registros de manera ascendente vamos a revisar si no ah borrado algun registro nulo. Nos devuelve las regiones con las cantidades de veces que se repite.

## Ordenar valores.
Para ordenar valores lo podemos hacer con el metodo sort_values y podemos ordenar tanto dataframes como series.
### Forma de uso:
#### Ordenar series:
```
dataframe.serie.sort_values()
```
#### Ordenar dataframe:
```
dataframe.sort_values()
```
### Parametros importantes.
#### Axis
```
sort_values(axis=0)
```
#### Ascending
```
sort_values(ascending=True)
```
#### Na_position
Por defecto van al final
```
sort_values(na_position='last')
```
#### Ejemplos:

````python
tabla.Vendedor.sort_values()
````

````python
tabla.sort_values(by=['Vendedor', 'Ganancia'])
````

## Condiciones booleanas.

Se utilizan vectores booleanos para filtrar información.

### Operadores

| Operador | Traducción | Simbolo |
|----------|----------|----------|
| AND    | Y   | &   |
| OR    | O   | |   |
| NOT    | No | ~   |

### Ejemplo

````python
tabla.Artículo == 'Gorras'
````

````python
tabla[tabla.Artículo == 'Gorras']
````

````python
tabla[(tabla.Artículo == 'Gorras') & (tabla.Región == 'Norte')]
````

## Manejo de strings

* serie.str.isnumeric()

### Ejemplos:

````python
tabla.Vendedor.str.contains('Trevor')
````

````python
tabla[tabla.Vendedor.str.contains('Trevor')]
````

````python
tabla[tabla.Vendedor.str.startswith('T')]
````
